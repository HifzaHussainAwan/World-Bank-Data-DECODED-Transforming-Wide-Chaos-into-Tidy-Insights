```python
#like now we are going to perform EDA(Exporlatory Data Analysis)
#First step of that is Univariant anaylsis 
#Second step is handling misiing values, feature engineering, dimensionality reduction,outliers removal, feature selection,Encoding,(test,train)splits 
#Third Step is to check Bivariant or Multivariant Analysis
```

```python
#Step one the Uni-variant Analysis to check the quality,outliers,missing values in data so it is going to perform on every feature in dataset
#1st we do uni-variant analysis for categoical 
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
sns.countplot(df['Sex'])
plt.title('The distribution of Sex using Count plot')
plt.show()
```

```python
#through this graph we can understand that the number or ratio of women is half of the men travelling in titanic 
```

```python
plt.figure(figsize=(10,30))
sns.countplot(df['Cabin'])
plt.title('The distribution of Cabins')
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.countplot(df['Embarked'])
plt.title('The distribution of Embarked')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Visualization of Survived')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='SibSp',data=df)
plt.title('Visualization of Siblings and Spous SibSp Feature  to detect Outliers')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Parch',data=df)
plt.title('Visualization of Parent and Child(Parch) feature to detect Outliers')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(
            x='Pclass', 
            data=df,         )

plt.title('Visualization of Pclass')

plt.show()
```

```python
# Now we do uni-variant analysis for numerical features
#first we are using boxplot to find the outliears in fare and age feautres
sns.boxplot(df['Age'])
plt.title('Distribution of Age to detect Outliers')
plt.show()
```

```python
sns.boxplot(df['Fare'])
plt.title('Distribution of Fare to detect Outliers using boxplot')
plt.show()
```

```python
sns.histplot(df['Fare'])
plt.title('Distribution of Fare to detect Outliers using histogram')
plt.show()
```

```python
#Bivarate Analysis
#I jsut want to check that what is the age of people who survived 
plt.figure(figsize=(10,3))  # 30 is too long for height, 6 is better
sns.scatterplot(x=df['Fare'], y=df['Age'])
plt.title('Age of People Who Survived')
plt.show()
```

```python
#To check that the number of women and men survived 
plt.figure(figsize=(8,3))  # 30 is too long for height, 6 is better
sns.barplot(x=df['Sex'], y=df['Survived'])
plt.title('Age of People Who Survived')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns

numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Plot correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.show()
```

```python
# Survived vs Age
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Survival Distribution by Age')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
# Survived vs Age
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Pclass', data=df)
plt.title('Survival Distribution by Age')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Pclass')
plt.show()
```

```python
# Survived vs Age
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Sex', data=df)
plt.title('Survival Distribution by Sex')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Sex')
plt.show()
```
