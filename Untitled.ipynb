```python
import seaborn as sns
import matplotlib.pyplot as plt

# The 'y' argument creates a horizontal bar plot, which is much better for many labels
plt.figure(figsize=(10, 70)) 
sns.countplot(y='Country Name', data=df)
plt.title('Count of Records per Country/Region')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 50))
sns.barplot(y='Country Name', x='1969', data=df)
plt.title('Population by Country (1969)')
plt.xlabel('Population')
plt.ylabel('Country Name')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 2))
sns.histplot(df['1969'], bins=5, kde=True)
plt.title('Population Distribution in 1969')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
